### 1.1 — Listas y operaciones básicas

Tienes la siguiente lista con la esperanza de vida de 10 países (en años).

````python
# Lista dada — NO modificar
esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]
````

**Ejercicio 1.1** — Usando Python puro (sin Pandas ni funciones built-in como `sum()` o `max()`):
- Calcula el **promedio** de la lista usando un bucle `for`
- Encuentra el **valor máximo** usando un bucle `for`
- Encuentra el **valor mínimo** usando un bucle `for`

````python
# Tu código aquí
suma = 0
items = 0
for x in esperanzas_de_vida:
    suma+=x
    items+=1
# Promedio
promedio = suma / items
print(f"El promedio de la lista esperanzas_de_vida es: {promedio:.2f} ")

# Máximo
valor_maximo = esperanzas_de_vida[0]

for x in esperanzas_de_vida:
    if x > valor_maximo:
        valor_maximo = x
print(f"El valor maximo de la lista es: {valor_maximo}")
    

# Mínimo
valor_minimo = esperanzas_de_vida[0]

for x in esperanzas_de_vida:
    if x < valor_minimo:
        valor_minimo = x

print(f"El valor minimo de la lista es: {valor_minimo}")
````

### 2.5 — Agrupaciones

🔧 **Ejercicio 2.5** — Usa `groupby()` para:
1. Calcular la **esperanza de vida promedio** por `region`
2. Calcular el **gasto en salud promedio** por `nivel_ingresos`
3. Calcular el **número de países** por región
4. ¿Qué región tiene la mayor esperanza de vida promedio? ¿Y la menor?

````python
# Esperanza de vida promedio por region
sm_limpio.groupby('region')['esperanza_vida'].mean()
````

````python
# gasto en salud promedio por nivel de ingreso
sm_limpio.groupby('nivel_ingresos')['gasto_salud_usd'].mean()
````

````python
# Calcular el numero de paises por region
sm_limpio.groupby('region')['pais'].count()
````

````python
# Que region tiene mayor y cual menor esperanza de vida promedio
vida_mayor = sm_limpio.groupby('region')['esperanza_vida'].mean().idxmax()
print(f"Region con mayor esperanza de vida: {vida_mayor}")

vida_menor = sm_limpio.groupby('region')['esperanza_vida'].mean().idxmin()
print(f"Region con menor esperanza de vida: {vida_menor}")
````
